```python
def reverse_string(string):
    return string[::-1]

def complement(string):
    complement = ""
    for char in string:
        if char == "a":
            complement += "t"
        elif char == "t":
            complement += "a"
        elif char == "g":
            complement += "c"
        elif char == "c":
            complement += "g"
    return complement

def complement_caps(string):
    complement = ""
    for char in string:
        if char == "A":
            complement += "T"
        elif char == "T":
            complement += "A"
        elif char == "G":
            complement += "C"
        elif char == "C":
            complement += "G"
    return complement
```

```python
def DNA_to_RNA(string):
    return string.replace("t", "u")
```
